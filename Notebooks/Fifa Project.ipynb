```python
df_team = pd.DataFrame()
sum = ['goals', 'assists', 'shots', 'shots_on_target','expected_goals_xg', 'expected_assists_xa','key_passes', 'successful_dribbles', 'dribbles_attempted','successful_passes', 'total_passes','crosses', 'successful_crosses','tackles', 'interceptions', 'clearances', 'blocks', 'recoveries', 'defensive_actions','aerial_duels_won', 'aerial_duels_lost','saves', 'clean_sheet', 'goals_conceded', 'penalty_saves','yellow_cards', 'red_cards', 'fouls_committed', 'fouls_suffered', 'offsides']
avg = ['pass_accuracy','save_percentage']
df_team['team'] = df['team'].unique()
for i in df[sum]:
    total = df.groupby('team')[i].sum()
    df_team[i] = df_team['team'].map(total)

for i in df[avg]:
    total_avg = df.groupby('team')[i].mean()
    df_team[i] = df_team['team'].map(total_avg)

```

```python
df_team = df_team.rename(columns = {'pass_accuracy': 'avg_pass_accuracy','save_percentage': 'avg_save_percentage',})
```

```python

df_team.info()
```

```python
numeric_df = df_team.select_dtypes(include = "number")
plt.figure(figsize=(16, 14))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={'shrink': 0.7})
plt.title('Correlation Heatmap of Team Stats')
plt.tight_layout()
plt.show()
```

```python
df_team = pd.DataFrame()
df_team["team"] = df['team'].unique()
number_of_players = df.groupby("team")["player_id"].nunique()
df_team["team_size"] = df_team["team"].map(number_of_players)
tvalue = df.groupby("team")["market_value_eur"].sum()
df_team["team_value"] = df_team["team"].map(tvalue)
goal_totals = df.groupby("team")["goals"].sum()
df_team["total_goals"] = df["team"].map(goal_totals)
expected_goals = df.groupby("team")["expected_goals_xg"]
gchance = (df.groupby("team")["goals"].sum()) / (df.groupby("team")["shots"].sum())
df_team["goal_chance"] = df_team["team"].map(gchance)
df_team


```

```python
plt.scatter(df_team["goal_chance"], df_team["team_value"])

plt.xlabel("Goal Chance")
plt.ylabel("Team Value")
plt.title("Relationship Between Team Value and Goal Chance")

plt.show()
```

```python
subset_cols = ['goals', 'dribbles_attempted', 'shots', 'key_passes',
               'successful_passes', 'recoveries', 'clearances', 'aerial_duels_won',
               'red_cards', 'clean_sheet', 'avg_save_percentage']

df_subset = df_team[subset_cols]
df_subset.head()
```

```python
pairs = [
    ('dribbles_attempted', 'goals'),
    ('key_passes', 'shots'),
    ('recoveries', 'successful_passes'),
    ('clearances', 'aerial_duels_won'),
    ('avg_save_percentage', 'red_cards'),
    ('avg_save_percentage', 'clean_sheet'),
]

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
for ax, (x, y) in zip(axes.flat, pairs):
    sns.regplot(data=df_team, x=x, y=y, ax=ax,
                scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
    ax.set_title(f'{x} vs {y}')

plt.tight_layout()
plt.show()
```
