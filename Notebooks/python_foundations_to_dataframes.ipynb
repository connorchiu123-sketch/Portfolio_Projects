## 1. Lists

A **list** is an ordered collection of values. You can put anything in a list, and you access items by their position (index), starting at `0`.

```python
# A simple list of player names
players = ["Cam Brennan", "Dre Marsh", "Mason Brennan", "Malik Townsend"]

print(players)
print("First player:", players[0])
print("Last player:", players[-1])
print("How many players:", len(players))
```

```python
# Slicing: grab a range of items [start:stop] (stop is NOT included)
print(players[0:2])   # first two players
print(players[1:])    # everything from index 1 onward

# Common list methods
players.append("Jordan Lee")   # add to the end
print(players)

players.sort()                 # sort alphabetically, in place
print(players)

players.remove("Dre Marsh")    # remove a specific value
print(players)
```

## 2. Dictionaries

A **dictionary** (`dict`) stores data as **key/value** pairs instead of positions. This is how you label data — much closer to how real-world records (like a single player) actually look.

```python
# A single player represented as a dictionary
player = {
    "name": "Cam Brennan",
    "position": "PF",
    "age": 21,
    "ppg": 7.6,
    "rpg": 4.5
}

print(player["name"])
print(player["ppg"])

# Add a new key, or update an existing one
player["team"] = "St. Louis Comets"
player["age"] = 22
print(player)
```

```python
# Useful dict methods
print(player.keys())
print(player.values())
print(player.items())

# A dictionary can hold ANOTHER dictionary as a value (nesting!)
player_nested = {
    "name": "Cam Brennan",
    "position": "PF",
    "stats": {
        "ppg": 7.6,
        "rpg": 4.5,
        "apg": 9.0
    }
}

# To reach a value inside the nested dict, chain the keys
print(player_nested["stats"]["ppg"])
```

## 3. For Loops

A **for loop** lets you repeat an action for every item in a collection (a list, a dict, etc.). This is how you process many records without writing repetitive code.

```python
# Looping over a list
players = ["Cam Brennan", "Mason Brennan", "Malik Townsend"]

for name in players:
    print("Player:", name)
```

```python
# enumerate() gives you the index AND the value while looping
for i, name in enumerate(players):
    print(i, "->", name)

# Looping over a dictionary with .items() gives you key AND value
player = {"name": "Cam Brennan", "position": "PF", "ppg": 7.6}

for key, value in player.items():
    print(key, ":", value)
```

## 4. Nested Data Structures (what a real API response looks like)

So far we've used one list, or one dict, at a time. Real data — like the JSON you'd get back from a sports API — is usually a **list of dictionaries**, and each dictionary can have **nested dictionaries** inside it.

Below is a mock API response for a team roster. This is intentionally shaped like what `requests.get(url).json()` would hand you in practice.

```python
api_response = {
    "team": "St. Louis Comets",
    "season": 2022,
    "players": [
        {
            "name": "Cam Brennan",
            "position": "PF",
            "age": 21,
            "stats": {"ppg": 7.6, "rpg": 4.5, "apg": 9.0, "per": 11.9}
        },
        {
            "name": "Dre Marsh",
            "position": "SG",
            "age": 36,
            "stats": {"ppg": 3.7, "rpg": 3.9, "apg": 7.6, "per": 11.4}
        },
        {
            "name": "Mason Brennan",
            "position": "PG",
            "age": 20,
            "stats": {"ppg": 16.7, "rpg": 1.8, "apg": 5.4, "per": 14.5}
        },
        {
            "name": "Malik Townsend",
            "position": "C",
            "age": 23,
            "stats": {"ppg": 8.5, "rpg": 5.7, "apg": 0.3, "per": 12.5}
        }
    ]
}

# The top level is a dict. "players" is a list. Each player is a dict.
print(type(api_response))
print(type(api_response["players"]))
print(type(api_response["players"][0]))
```

```python
# To get to any individual stat, you chain: dict key -> list index -> dict key -> dict key
first_player_ppg = api_response["players"][0]["stats"]["ppg"]
print("First player's ppg:", first_player_ppg)

# Looping through a nested structure is the normal way to pull data out of it
for player in api_response["players"]:
    name = player["name"]
    position = player["position"]
    ppg = player["stats"]["ppg"]
    print(f"{name} ({position}): {ppg} points per game")
```

```python
# You can build a new list out of a loop over nested data
high_scorers = []

for player in api_response["players"]:
    if player["stats"]["ppg"] > 8:
        high_scorers.append(player["name"])

print("Players averaging more than 8 ppg:", high_scorers)
```

## 5. From Nested Data to a DataFrame

Looping through nested dicts/lists by hand works, but it gets tedious once you have hundreds or thousands of records. That's exactly the problem **pandas** solves. A DataFrame is essentially a table built out of exactly the kind of structure we just worked with.

Below, `pd.DataFrame()` converts the `players` list from our mock API response straight into a table.

```python
import pandas as pd

# pandas can't flatten the nested "stats" dict automatically, so let's flatten it ourselves first
flat_players = []

for player in api_response["players"]:
    row = {
        "name": player["name"],
        "position": player["position"],
        "age": player["age"],
        "ppg": player["stats"]["ppg"],
        "rpg": player["stats"]["rpg"],
        "apg": player["stats"]["apg"],
        "per": player["stats"]["per"]
    }
    flat_players.append(row)

df_mock = pd.DataFrame(flat_players)
df_mock
```

## Recap

- **Lists** hold ordered items, accessed by position.
- **Dicts** hold labeled items, accessed by key — and can be nested inside each other.
- **For loops** let you process every item in a list or dict without repeating code.
- Real-world data (like an API response) is usually **nested**: a list of dicts, sometimes with dicts inside those dicts.
- **pandas** turns that same shape of data — or a CSV/Excel file — into a **DataFrame** you can inspect with `.head()`, `.info()`, `.describe()`, filter with boolean conditions, and summarize with things like `.value_counts()`.
